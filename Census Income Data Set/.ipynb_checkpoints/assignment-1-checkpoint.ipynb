```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
import time
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

warnings.filterwarnings("ignore")

# Load data
columns_name = ['age', 'workclass', 'fnlwgt', 'education', 'education_num', 'marital_status',
                'occupation', 'relationship', 'race', 'sex', 'capital_gain', 'capital_loss',
                'hours_perweek', 'native_country', 'income']

df_train = pd.read_csv("C:/Users/wsy/Downloads/Census Income Data Set/Census Income Data Set/adult.data", names=columns_name)
df_test = pd.read_csv("C:/Users/wsy/Downloads/Census Income Data Set/Census Income Data Set/adult.test", names=columns_name, skiprows=1)

# Handle missing values
# Replace missing values with NaN and drop rows with missing values to ensure data integrity
df_train = df_train.replace(' ?', np.nan).dropna()
df_test = df_test.replace(' ?', np.nan).dropna()

# Feature engineering
# Combine datasets for uniform processing
df = pd.concat([df_train, df_test], axis=0)

# Encode categorical variables
# Convert categorical variables to numerical values for model compatibility
for column in df.select_dtypes(include=['object']).columns:
    if column != 'income':
        le = LabelEncoder()
        df[column] = le.fit_transform(df[column])

# Split the combined dataset back into training and testing sets
df_train = df.iloc[:len(df_train), :]
df_test = df.iloc[len(df_train):, :]

# Standardize numerical features
# Standardize numerical features to improve model performance
scaler = StandardScaler()
numerical_features = df.select_dtypes(include=['int64', 'float64']).columns
df_train[numerical_features] = scaler.fit_transform(df_train[numerical_features])
df_test[numerical_features] = scaler.transform(df_test[numerical_features])

# Prepare training and testing data
X_train = df_train.drop('income', axis=1)
y_train = df_train['income'].apply(lambda x: 1 if x == ' >50K' else 0)

X_test = df_test.drop('income', axis=1)
y_test = df_test['income'].apply(lambda x: 1 if x == ' >50K.' else 0)

# Train and evaluate models
# Logistic Regression
start_time = time.time()
lr = LogisticRegression()
lr.fit(X_train, y_train)
train_time_lr = time.time() - start_time

start_time = time.time()
y_pred_lr = lr.predict(X_test)
predict_time_lr = time.time() - start_time

print("Logistic Regression Accuracy:", accuracy_score(y_test, y_pred_lr))
print("Logistic Regression Classification Report:\n", classification_report(y_test, y_pred_lr))

# Support Vector Machine
start_time = time.time()
svm = SVC()
svm.fit(X_train, y_train)
train_time_svm = time.time() - start_time

start_time = time.time()
y_pred_svm = svm.predict(X_test)
predict_time_svm = time.time() - start_time

print("SVM Accuracy:", accuracy_score(y_test, y_pred_svm))
print("SVM Classification Report:\n", classification_report(y_test, y_pred_svm))

# Plot confusion matrix
def plot_confusion_matrix(y_true, y_pred, title):
    cm = confusion_matrix(y_true, y_pred)
    plt.figure(figsize=(8, 6))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=['<=50K', '>50K'], yticklabels=['<=50K', '>50K'])
    plt.xlabel('Predicted')
    plt.ylabel('Actual')
    plt.title(title)
    plt.show()

plot_confusion_matrix(y_test, y_pred_lr, 'Logistic Regression Confusion Matrix')
plot_confusion_matrix(y_test, y_pred_svm, 'SVM Confusion Matrix')

# Data visualization
# Age distribution
plt.figure(figsize=(10, 6))
sns.histplot(df_train['age'], bins=30, kde=True)
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()

# Income distribution
plt.figure(figsize=(10, 6))
sns.countplot(x='income', data=df_train)
plt.title('Income Distribution')
plt.xlabel('Income')
plt.ylabel('Count')
plt.show()

# Education level vs income
plt.figure(figsize=(14, 8))
sns.countplot(x='education', hue='income', data=df_train)
plt.title('Education Level vs Income')
plt.xlabel('Education Level')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()

# Data visualization for Age vs Income
plt.figure(figsize=(10, 6))
sns.boxplot(x=y_test, y=df_test['age'])
plt.title('Age vs Income')
plt.xlabel('Income')
plt.ylabel('Age')
plt.show()

# Data visualization for Workclass vs Income
plt.figure(figsize=(10, 6))
sns.countplot(x='workclass', hue=y_test, data=df_test)
plt.title('Workclass vs Income')
plt.xlabel('Workclass')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
# Print training and prediction time
print(f"Logistic Regression Training Time: {train_time_lr:.4f} seconds")
print(f"Logistic Regression Prediction Time: {predict_time_lr:.4f} seconds")
print(f"SVM Training Time: {train_time_svm:.4f} seconds")
print(f"SVM Prediction Time: {predict_time_svm:.4f} seconds")
```
