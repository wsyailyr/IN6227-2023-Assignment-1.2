```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load data
columns_name = ['age', 'workclass', 'fnlwgt', 'education', 'education_num', 'marital_status',
                'occupation', 'relationship', 'race', 'sex', 'capital_gain', 'capital_loss',
                'hours_perweek', 'native_country', 'income']

df_train = pd.read_csv("C:/Users/wsy/Downloads/Census Income Data Set/Census Income Data Set/adult.data", names=columns_name)
df_test = pd.read_csv("C:/Users/wsy/Downloads/Census Income Data Set/Census Income Data Set/adult.test", names=columns_name, skiprows=1)

# Combine datasets for visualization
df = pd.concat([df_train, df_test], axis=0)

# Visualize each column
plt.figure(figsize=(20, 15))

# Age distribution
plt.subplot(3, 3, 1)
sns.histplot(df['age'], bins=30, kde=True)
plt.title('Age Distribution')

# Workclass distribution
plt.subplot(3, 3, 2)
sns.countplot(y='workclass', data=df, order=df['workclass'].value_counts().index)
plt.title('Workclass Distribution')

# fnlwgt distribution
plt.subplot(3, 3, 3)
sns.histplot(df['fnlwgt'], bins=30, kde=True)
plt.title('Fnlwgt Distribution')

# Education distribution
plt.subplot(3, 3, 4)
sns.countplot(y='education', data=df, order=df['education'].value_counts().index)
plt.title('Education Distribution')

# Education.num distribution
plt.subplot(3, 3, 5)
sns.histplot(df['education_num'], bins=30, kde=True)
plt.title('Education_num Distribution')

# Marital.status distribution
plt.subplot(3, 3, 6)
sns.countplot(y='marital_status', data=df, order=df['marital_status'].value_counts().index)
plt.title('Marital.status Distribution')

# Occupation distribution
plt.subplot(3, 3, 7)
sns.countplot(y='occupation', data=df, order=df['occupation'].value_counts().index)
plt.title('Occupation Distribution')

# Relationship distribution
plt.subplot(3, 3, 8)
sns.countplot(y='relationship', data=df, order=df['relationship'].value_counts().index)
plt.title('Relationship Distribution')

# Race distribution
plt.subplot(3, 3, 9)
sns.countplot(y='race', data=df, order=df['race'].value_counts().index)
plt.title('Race Distribution')

plt.tight_layout()
plt.show()

plt.figure(figsize=(20, 15))

# Sex distribution
plt.subplot(3, 3, 1)
sns.countplot(y='sex', data=df, order=df['sex'].value_counts().index)
plt.title('Sex Distribution')

# Capital.gain distribution
plt.subplot(3, 3, 2)
sns.histplot(df['capital_gain'], bins=30, kde=True)
plt.title('Capital_gain Distribution')

# Capital.loss distribution
plt.subplot(3, 3, 3)
sns.histplot(df['capital_loss'], bins=30, kde=True)
plt.title('Capital.loss Distribution')

# Hours.per.week distribution
plt.subplot(3, 3, 4)
sns.histplot(df['hours_perweek'], bins=30, kde=True)
plt.title('Hours_perweek Distribution')

# Native.country distribution
plt.subplot(3, 3, 5)
sns.countplot(y='native_country', data=df, order=df['native_country'].value_counts().index)
plt.title('Native_country Distribution')

# Income distribution
plt.subplot(3, 3, 6)
sns.countplot(y='income', data=df, order=df['income'].value_counts().index)
plt.title('Income Distribution')

plt.tight_layout()
plt.show()
```
